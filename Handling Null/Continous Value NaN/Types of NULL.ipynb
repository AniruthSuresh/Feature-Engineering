### 1. MCAR : Missing Completely At Random

 When data is MCAR, there is absolutely no relationship between the data missing and any other values, observed or missing, within the dataset.

```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
df.head(5)

# embarked -> where did they board from : Southampton(S) , Cherbourg(C) , Queenstown(Q)
```

```python
df.isnull().sum()
```

```python
null_values = df.isnull().sum()

# Print the columns with null values and their respective counts
columns_with_nulls = null_values[null_values > 0]
if not columns_with_nulls.empty:
    print("Columns with Null Values:")
    for column, count in columns_with_nulls.items():
        print(f"{column}: {count}")
else:
    print("No Null Values Found in Any Column.")
```

### 2. MNAR : Missing Not At Random

There is absolutely some relationship between the data missing and any other values, observed or missing, within the dataset.

```python
import numpy as np

# df[df["Cabin"].isnull()]
df["cabin-null"] = np.where(df["Cabin"].isnull(),1,0)
print(f"mean is {df['cabin-null'].mean()}")
df.sample(10)

```

```python
df.groupby(['Survived'])['cabin-null'].mean()  # clearly there is a relationship between survival and the cabin being nu;;
```

### 3. MAR : Missing At Random

This means that the probability of a data point being missing is unrelated to the value of the data point, but it may be related to other variables in the datasetm
