### METHOD 4 : END OF DISTRIBUTION IMPLEMENTATION

```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv",usecols = ['Age',"Fare",'Survived'])
df.isnull().sum()

# print(df.shape)
```

```python
# so basically , replace the NAN values with the values which are in the end of the distribution
from matplotlib import pyplot as plt

extreme = df["Age"].mean() + 3*df["Age"].std()
print(extreme)

plt.hist(df.Age , bins = 20 , rwidth = 0.8)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
def impute_NAN(df,variable,median,extreme):
    df[variable+"_end_dsit"] = df[variable].fillna(extreme) # fillna -> replace the NULL values with the specified value
    df[variable].fillna(median,inplace = True)
    return df
```

```python
impute_NAN(df,"Age",df["Age"].median(),extreme)
```

```python
plt.hist(df.Age_end_dsit , bins = 20 , rwidth = 0.8)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df.Age , bins = 20 , rwidth = 0.8)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Kernet Density Estimators !
from matplotlib import pyplot as plt


fig = plt.figure()
ax = fig.add_subplot(111)
df.Age.plot(kind='kde', ax=ax, color='red')
df.Age_end_dsit.plot(kind='kde', ax=ax, color='black')

lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')


# clearly the distribution is very close to the variable !! -> almost no change 
```
